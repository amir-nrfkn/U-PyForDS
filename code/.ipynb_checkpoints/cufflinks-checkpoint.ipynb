# Plotly and Cufflinks

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
from plotly import __version__
```

```python
print(__version__)
```

```python
import cufflinks as cf
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode,plot,iplot
```

```python
init_notebook_mode(connected=True)
# plotly connects pandas and python to interactive js library
```

```python
cf.go_offline()
# allows cufflinks to work offline
```

```python
# DATA
df = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,71]})
```

```python
# pandas plot
df.plot()
```

```python
# plotly interactive image
df.iplot()
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers', size=5)
```

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
# bar plot for this type of dataset needs some aggregate function to make sense, like sum or count
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

## 3D surface plot

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
```

```python
df3
```

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
df.iplot(kind='hist', bins=40)
```

```python
df[['A','B']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble', x='A', y='B', size='C')
```

```python
# columns must be numeric
df.scatter_matrix()
```
