```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,31,12)
y = (x**2)
```

```python
x
```

```python
y
```

## Functional Method for Plotting

```python
plt.plot(x, y, 'b-')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

## OOP Method for Plots

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')
```

### first method (more manual)

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.4, 0.3, 0.4])

axes1.plot(x, y)
axes1.set_title('large')
axes2.plot(y, x)
axes2.set_title('small')
```

### second method (subplots arranged by rows and columns)

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()

# iter over axes object
# for current_ax in axes:
#     current_ax.plot(y,x)

# can index axes object too
axes[0].plot(x, y)
axes[0].set_title('first plot xy')

axes[1].plot(y, x)
axes[1].set_title('second plot yx')

plt.tight_layout()
```

```python
# axes object is an array of matplotlib axes (list of axes objects) can iter over it
axes
```
