```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,31,12)
y = (x**2)
```

```python
x
```

```python
y
```

## Functional Method for Plotting

```python
plt.plot(x, y, 'b-')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

## OOP Method for Plots

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.2, 0.3, 0.4])

axes1.plot(x, y)
axes1.set
axes2.plot(y, x)
axes2.set_title('small')
```
